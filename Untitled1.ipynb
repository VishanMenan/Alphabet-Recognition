```python
import os

# Path to your merged dataset's test folder
merged_test_path = r'C:\Users\HP\Documents\Python Project\MergedDataset\train'

# List all files and directories in the test folder
print("Contents of the train folder:")
for item in os.listdir(merged_test_path):
    print(item)
```

```python
# pixels range from 0 to 255 so dividing by 255 so value lie between 0 and 1
train_datagen = ImageDataGenerator(rescale=1./255)
train = train_datagen.flow_from_directory(
    directory = r'C:\Users\HP\Documents\Python Project\MergedDataset\train',
    target_size=(300, 300))
```

```python
val = train_datagen.flow_from_directory(
    directory=r'C:\Users\HP\Documents\Python Project\MergedDataset\validation',
    target_size=(300, 300),
    shuffle=False
)
```

```python
test = train_datagen.flow_from_directory(
    directory = r'C:\Users\HP\Documents\Python Project\MergedDataset\test',
    target_size=(300, 300))
```

```python
print("Class distribution in train:")
print(train.class_indices)

print("Class distribution in validation:")
print(val.class_indices)

print("Class distribution in test:")
print(test.class_indices)
```

```python
import tensorflow as tf

model = tf.keras.models.Sequential([
    # Add an explicit Input layer
    tf.keras.layers.Input(shape=(300, 300, 3)),

    # The first convolution
    tf.keras.layers.Conv2D(16, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    # The second convolution
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    # The third convolution
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    # The fourth convolution
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    # The fifth convolution
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    # Flatten the results to feed into a Dense layer
    tf.keras.layers.Flatten(),

    # Fully connected layer with 512 neurons
    tf.keras.layers.Dense(512, activation='relu'),

    # Output layer with 52 neurons for 52 classes
    tf.keras.layers.Dense(52, activation='softmax')
])
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
from tqdm import tqdm
# extracting values from val test generator
val.reset()
X_val,y_val = next(val)
for i in tqdm(range(int(len(val)/32)-1)): #1st batch is already fetched before the for loop.
    img, label = next(val)
    X_val = np.append(X_val, img, axis=0 )
    y_val = np.append(y_val, label, axis=0)
print(X_val.shape, y_val.shape)
```

```python
checkpointer = EarlyStopping(monitor = 'val_accuracy', verbose = 1, restore_best_weights=True,mode="max",patience = 15)
# checkpointer to monitor accuracy and preventing overfit
```

```python
history = model.fit(
            train ,
            steps_per_epoch =len(train)//32, #batch size is 32
            epochs=15,
            verbose=1,
            validation_data=(X_val,y_val),
            callbacks = [checkpointer])
```

```python
training_loss = history.history['loss']
val_loss = history.history['val_loss']
plt.rcParams['figure.figsize'] = [10, 5]
plt.style.use(['default'])
# Create count of the number of epochs
epoch_count = range(1, len(training_loss) + 1)

# Visualize loss history
plt.plot(epoch_count, training_loss, 'r--')
plt.plot(epoch_count, val_loss, 'b-')
plt.legend(['Training Loss', 'Val Loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

```python
training_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']

# Create count of the number of epochs
epoch_count = range(1, len(training_accuracy) + 1)

# Visualize Accuracy history
plt.plot(epoch_count, training_accuracy, 'r--')
plt.plot(epoch_count, val_accuracy, 'b-')
plt.legend(['Training Accuracy', 'Val Accuracy'])
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim(top = 1)
plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import PIL
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (BatchNormalization, Conv2D, MaxPooling2D, Activation, Flatten, Dropout, Dense)
from tensorflow.keras import backend as K
```

```python
train_path = r'C:\Users\HP\Documents\Python Project\MergedDataset\train'
val_path = r'C:\Users\HP\Documents\Python Project\MergedDataset\validation'
test_path = r'C:\Users\HP\Documents\Python Project\MergedDataset\test'

# Verify directories
print("Training directory contents:", os.listdir(train_path))
print("Validation directory contents:", os.listdir(val_path))
print("Test directory contents:", os.listdir(test_path))
```

```python
# Normalize images
train_datagen = ImageDataGenerator(rescale=1./255)

train = train_datagen.flow_from_directory(
    directory=train_path,
    target_size=(300, 300),
    batch_size=32
)
val = train_datagen.flow_from_directory(
    directory=val_path,
    target_size=(300, 300),
    batch_size=32
)
test = train_datagen.flow_from_directory(
    directory=test_path,
    target_size=(300, 300),
    batch_size=32,
    shuffle=False  # Ensure labels are consistent for evaluation
)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=(300, 300, 3)),
    tf.keras.layers.Conv2D(16, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(52, activation='softmax')
])

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```
